**unique() method**

In pandas, the **unique()** method is used to get an array of unique values in a pandas Series or DataFrame column.

For a DataFrame:if you call data.unique(), it returns an array containing all unique values present in the entire DataFrame, treating the DataFrame as if it were a single one-dimensional array. This means it will flatten the DataFrame and return unique values from all columns, preserving the order in which they appear.

```python
label_encoder = preprocessing.LabelEncoder()

non_numerical_columns = [col for col in df.columns if (df[col].dtype == 'object' or df[col].dtype == 'bool')]

for column in non_numerical_columns:
    print(f"Unique values for column '{column}':")
    print(df[column].unique())
    print()
```

Getting the list of non-numerical columns

by **.unique** command we can see what categories we have in each non numerical column.

```python
pd.set_option('future.no_silent_downcasting', True)

mapping_embarked = {'Southampton': int(0), 'Cherbourg': int(1), 'Queenstown': int(2)}
df['embarked'] = df['embarked'].replace(mapping_embarked)

mapping_embark_town = {'Southampton': int(0), 'Cherbourg': int(1), 'Queenstown': int(2)}
df['embark_town'] = df['embark_town'].replace(mapping_embark_town)

mapping_deck = {'C' : 0, 'E' : 1, 'G':2 ,'D':3 ,'A':4 ,'B':5 ,'F':6}
df['deck'] = df['deck'].replace(mapping_deck)

mapping_calss = {'Third':3 ,'First':1 ,'Second':2}
df['class'] = df['class'].replace(mapping_calss)

for column in non_numerical_columns:
    print(f"Unique values for column '{column}':")
    if column not in ['embarked', 'embark_town', 'deck', 'class']:
        df[column] = label_encoder.fit_transform(df[column])

    print(df[column].unique())
    print()
```

**Warning Handeling**:

This code sets the pandas option ‘future.no_silent_downcasting’ to True at the beginning, which opts-in to the future behavior of replace and avoids the warning.

**Code Explanation**:

The code performs categorical encoding to convert non-numeric data into numeric representations, which is often necessary for machine learning algorithms.
Mappings are provided to ensure consistent encoding of categorical values across different columns.
For columns explicitly included in the mappings, values are replaced directly based on the predefined mappings.
For other columns, label encoding is used, which assigns unique integer labels to each unique category in the column.

**Mapping Dictionaries**:

mapping_embarked, mapping_embark_town, mapping_deck, and mapping_calss are dictionaries containing mappings of categorical values to numerical representations for the corresponding columns in the DataFrame.

**Replacing Values:**

The .replace() method is used to replace values in the DataFrame columns according to the mappings provided in the dictionaries.
For example, df['embarked'] = df['embarked'].replace(mapping_embarked) replaces the categorical values in the 'embarked' column with their corresponding numerical representations based on the mapping_embarked dictionary.

**Label Encoding:**

For columns not explicitly included in the mappings ('embarked', 'embark_town', 'deck', and 'class'), label encoding is performed using scikit-learn's LabelEncoder.
Label encoding is a method of converting categorical values into numerical representations. It assigns a unique integer to each unique category in the column.
The fit_transform() method of LabelEncoder is used to fit the encoder to the column data and transform the categories into numerical labels simultaneously.
This transformation replaces the original categorical values in the DataFrame with their corresponding numerical labels.

```python
def fit_normal_distribution(df, variable, method_name, color):
    # Fit a normal distribution to the df
    mu, std = norm.fit(df[variable].dropna())
    
    # Plot KDE of the df
    df[variable].plot(kind='kde', color=color, label=method_name)
    
    # Plot the fitted normal distribution
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, color=color, linestyle='--', linewidth=2)

# Function to calculate mean and median, and fill missing values
def mean_median(df, variable):
    df[variable+'_mean']  = df[variable].fillna(df[variable].mean())
    df[variable+'_median']  = df[variable].fillna(df[variable].median())

# Function to fill missing values with random samples from non-missing values
def age_fill(df, variable):
    df[variable+'_random'] = df[variable]
    random_values = df[variable].dropna().sample(df[variable].isnull().sum(), random_state=0)
    random_values.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random'] = random_values

# Function to fill missing values with extreme values
def End_distribution(df, variable, extreme):
    df[variable+'_end_distribution'] = df[variable].fillna(extreme)

# Apply filling methods
mean_median(df, 'age')
age_fill(df, 'age')
extreme = (df['age'].mean() + (3 * df['age'].std()))
End_distribution(df, 'age', extreme)

# Plotting KDE and fitted normal distribution for each method
plt.figure(figsize=(12,8))
fit_normal_distribution(df, 'age', 'Original', 'b')
fit_normal_distribution(df, 'age_mean', 'Mean Fill', 'g')
fit_normal_distribution(df, 'age_median', 'Median Fill', 'grey')
fit_normal_distribution(df, 'age_end_distribution', 'End Distribution Fill', 'y')
fit_normal_distribution(df, 'age_random', 'Random Fill', 'r')


plt.legend()
plt.title('Comparison of Filled Data with Normal Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

def predict_age(df, algorithm, features):
    # Split the dataset into two parts: one with missing 'age' values and one without
    missing_age_data = df[df['age'].isnull()]
    complete_age_data = df.dropna(subset=['age'])
    df[f'predicted_age_{algorithm}'] = df['age']
    # Select features and target variable
    X_train = complete_age_data[features]  # Features
    y_train = complete_age_data['age']  # Target variable
    X_test = missing_age_data[features]  # Features for which we want to predict age
    
    # Initialize model based on the selected algorithm
    if algorithm == 'RandomForest':
        model = RandomForestRegressor(random_state=42)
    elif algorithm == 'SVR':
        model = SVR()
    elif algorithm == 'MLP':
        model = MLPRegressor(random_state=42)
    elif algorithm == 'KNN':
        model = KNeighborsRegressor()
    elif algorithm == 'LinearRegression':
        model = LinearRegression()
    elif algorithm == 'XGBoost':
        model = XGBRegressor(random_state=42)
    elif algorithm == 'GaussianProcess':
        kernel = RBF()
        model = GaussianProcessRegressor(kernel=kernel)
    else:
        raise ValueError("Invalid algorithm name.")
    
    # Train the machine learning model
    model.fit(X_train, y_train)
    
    # Predict ages
    predicted_ages = model.predict(X_test)
    
    # Round predicted ages to the nearest integer
    predicted_ages_rounded = [round(age) for age in predicted_ages]
    
    # Convert predicted ages to integers
    predicted_ages_int = [int(age) for age in predicted_ages_rounded]
    
    # Add predicted ages as new column in the dataframe
    df.loc[df[f'predicted_age_{algorithm}'].isnull(), f'predicted_age_{algorithm}'] = predicted_ages_int
    df[f'predicted_age_{algorithm}'] = df[f'predicted_age_{algorithm}'].astype(int)
    
    return df

# Assuming 'df' is your DataFrame
# Define features to be used for prediction
features = ['who', 'sibsp', 'class']

# Apply RandomForest algorithm
df = predict_age(df, 'RandomForest', features)

# Apply SVR algorithm
df = predict_age(df, 'SVR', features)

# Apply MLP algorithm
df = predict_age(df, 'MLP', features)

# Apply KNN algorithm
df = predict_age(df, 'KNN', features)

# Apply LinearRegression algorithm
df = predict_age(df, 'LinearRegression', features)

# Apply XGBoost algorithm
df = predict_age(df, 'XGBoost', features)

# Apply GaussianProcess algorithm
df = predict_age(df, 'GaussianProcess', features)

# Now, 'df' contains predicted ages from all algorithms as new columns for comparison


import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame containing predicted ages from different algorithms
# Columns 'age_original', 'predicted_age_RandomForest', 'predicted_age_SVR', etc. contain predicted ages

# List of column names containing predicted ages from different algorithms
predicted_age_columns = ['age','predicted_age_RandomForest', 'predicted_age_SVR', 'predicted_age_MLP', 'predicted_age_KNN', 'predicted_age_LinearRegression', 'predicted_age_XGBoost', 'predicted_age_GaussianProcess']

# Plot KDEs for each predicted age column
plt.figure(figsize=(12, 8))
for column in predicted_age_columns:
    # Use sns.kdeplot() to plot KDE
    sns.kdeplot(df[column], label=column)

plt.title('Kernel Density Estimate of Predicted Ages from Different Algorithms')
plt.xlabel('Predicted Age')
plt.ylabel('Density')
plt.legend()
plt.show()

```

Using the mean to calculate the survival proportion makes sense when dealing with a binary variable like 'Survived'

**Binary nature of 'Survived' column:** The 'Survived' column typically contains binary values where 0 represents "not survived" and 1 represents "survived". Since it's a binary variable, taking the mean of this column essentially gives us the proportion of passengers who survived. For example, if 60% of passengers in a particular group survived, the mean of the 'Survived' column for that group would be 0.6.

**Interpretation as a proportion:** By taking the mean of a binary variable, we can interpret the result as a proportion or percentage. For instance, if the mean of the 'Survived' column is 0.6 for a group, it means that 60% of passengers in that group survived.

**Consistency with mathematical representation:** Mathematically, the mean of a binary variable corresponds to the proportion of "1" values in that variable. In the context of survival analysis, "1" typically represents survival, so taking the mean aligns with this interpretation.

Plot a grouped bar chart using Matplotlib to show the average fare paid by passengers,
grouped by their class and survival status.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

grouped = df.groupby(['class', 'survived'])['fare'].mean().unstack()

fig, ax = plt.subplots(figsize=(10,6))
width = 0.35 
labels = grouped.index  
x = np.arange(len(labels))  

rects1 = ax.bar(x - width/2, grouped[0], width, label='Did not survive')
rects2 = ax.bar(x + width/2, grouped[1], width, label='Survived')

ax.set_xlabel('Class')
ax.set_ylabel('Average Fare')
ax.set_title('Average Fare by Class and Survival Status')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

**Explanation of the figure**

The bar chart shows the average fare paid by passengers in three different classes who either survived (blue bars) or did not survive (orange bars).

- **First Class:** The average fare for survivors is higher than for non-survivors. This could suggest that within the first class, passengers who paid more (perhaps for better cabins or services) had a higher chance of survival.

- **Second and Third Classes:** The average fares for survivors and non-survivors are very similar. This suggests that in these classes, the fare paid did not significantly affect the chances of survival.

- **Fare Disparity:** There's a clear disparity in fares between the classes. First-class passengers, both survivors and non-survivors, paid significantly more than passengers in the second and third classes. This reflects the socio-economic differences among the passengers.
